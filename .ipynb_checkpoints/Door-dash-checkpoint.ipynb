```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('historical_data.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['actual_delivery_time'] = pd.to_datetime(df['actual_delivery_time'])
```

```python
df.info()
```

```python
# creating the target column

t_delta = []

for i in range(len(df)):
    t1 = df['created_at'][i]
    t2 = df['actual_delivery_time'][i]
    dif = pd.Timedelta(t2 - t1).seconds
    t_delta.append(dif)
```

```python
df['actual_time_taken'] = t_delta
```

```python
df.head()
```

```python
store_counts = df['store_id'].value_counts().to_dict() # too many unique values, performing encoding
stores = []

for i in range(len(df)):
    freq = store_counts[df['store_id'][i]]
    if freq < 100:
        stores.append(0)
    elif freq >= 100 and freq < 200:
        stores.append(1)
    elif freq >= 200 and freq < 300:
        stores.append(2)
    else:
        stores.append(3)
        
df['store_id_encoded'] = stores
```

```python
df['store_id_encoded'].value_counts()
```

```python
df.head()
```

```python
df['store_primary_category'].value_counts() # too many categories, conflicting categories, dropping

df.drop(['store_primary_category'], axis=1, inplace=True)
```

```python
# also dropping created_at , actual_delivery_time and store_id since we don't need them anymore

df.drop(['created_at','actual_delivery_time','store_id'], axis=1, inplace=True)
```

```python
df.duplicated().sum() # No duplicate rows
```

```python
df.describe()
```

```python
# min of subtotal can't be zero

ind1 = df[df['subtotal'] == 0].index # dropping these rows
df.drop(ind1, inplace=True)
```

```python
# min_item_price can't be -86

ind2 = df[df['min_item_price'] < 0].index # min price can be zero for free items, deleting rows with  negative values
df.drop(ind2, inplace=True)
```

```python
# max_item_price zero

df[df['max_item_price'] == 0] # freebies i guess
```

```python
# onshift_dashers, busy_dashers and outstanding_orders can't be zero

ind3 = df[df['total_onshift_dashers'] < 0].index # dropping
df.drop(ind3, inplace=True)
```

```python
ind4 = df[df['total_busy_dashers'] < 0].index # dropping
df.drop(ind4, inplace=True)
```

```python
ind5 = df[df['total_outstanding_orders'] < 0].index # dropping
df.drop(ind5, inplace=True)
```

```python
# checking to see if total_busy_dashers > total_onshift_dashers, since the former should be a subset of the later

ind6 = df[df['total_busy_dashers'] > df['total_onshift_dashers']].index # way too many rows, dropping all
df.drop(ind6, inplace=True)
```

```python
ind7 = df[df['actual_time_taken'].isna()].index # dropping null values from target col
df.drop(ind7, inplace=True)
```

```python
# train-test-split

x = df.drop('actual_time_taken', axis=1)
y = df['actual_time_taken']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=56)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
x_train.head()
```

```python
# filling null values now

print(x_train.isnull().sum())
x_train.isnull().mean()*100
```

```python
# market_id and order_protocol are categorical so using most_frequent strat from simpleimputer

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
x_train[['market_id','order_protocol']] = imputer.fit_transform(x_train[['market_id','order_protocol']])
x_test[['market_id','order_protocol']] = imputer.transform(x_test[['market_id','order_protocol']])
```

```python
# filling the rest of null values with KNNimputer

from sklearn.impute import KNNImputer

kimp = KNNImputer()

kimp.fit(x_train)

x_train2 = kimp.transform(x_train)
x_test2 = kimp.transform(x_test)

x_train2 = pd.DataFrame(x_train2, columns=x_train.columns)
x_test2 = pd.DataFrame(x_test2, columns=x_test.columns)
```

```python
x_train2.head()
```

```python
x_train2.info()
```

```python
# Correlation

plt.figure(figsize=(17,13))
sns.heatmap(x_train2.corr(), annot = True)
plt.show()
```

```python
# total_onshift_dashers, total_busy_dashers and total_outstanding_orders are all showing high collinearity
# Printing VIF values 

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Variables'] = x_train2.columns
vif['VIF'] = [variance_inflation_factor(x_train2.values, i) for i in range(x_train2.shape[1])]

vif
```

```python
# dropping total_busy_dashers because of highest vif value

x_train2.drop(['total_busy_dashers'], axis=1, inplace=True)
x_test2.drop(['total_busy_dashers'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(17,13))
sns.heatmap(x_train2.corr(), annot = True)
plt.show()
```

```python
vif = pd.DataFrame()
vif['Variables'] = x_train2.columns
vif['VIF'] = [variance_inflation_factor(x_train2.values, i) for i in range(x_train2.shape[1])]

vif
```

```python
# dropping total_onshift_dashers

x_train2.drop(['total_onshift_dashers'], axis=1, inplace=True)
x_test2.drop(['total_onshift_dashers'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(17,13))
sns.heatmap(x_train2.corr(), annot = True)
plt.show()
```

```python
# base model
import xgboost as xg
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor

lr = LinearRegression()
dt = DecisionTreeRegressor(random_state=56)
knn = KNeighborsRegressor()
rf = RandomForestRegressor(random_state=56)
gb = GradientBoostingRegressor(random_state=56)
xgb = xg.XGBRegressor(objective ='reg:linear', verbosity=0, random_state=56)

models = [lr,dt,knn,rf,gb,xgb]

for m in models:
    m.fit(x_train2, y_train)
    train_pred = m.predict(x_train2)
    test_pred = m.predict(x_test2)
    print(m, ':', np.sqrt(mse(train_pred, y_train)).round(4), np.sqrt(mse(test_pred, y_test)).round(4))
```

```python
# Standard scaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

scaler = StandardScaler()
scaler2 = MinMaxScaler()

scaler.fit(x_train2)

x_train_scaled = scaler.transform(x_train2)
x_test_scaled = scaler.transform(x_test2)

x_train2 = pd.DataFrame(x_train_scaled, columns=x_train2.columns)
x_test2 = pd.DataFrame(x_test_scaled, columns=x_test2.columns)

x_train2.head()
```

```python
for m in models:
    m.fit(x_train2, y_train)
    train_pred = m.predict(x_train2)
    test_pred = m.predict(x_test2)
    print(m, ':', np.sqrt(mse(train_pred, y_train)).round(4), np.sqrt(mse(test_pred, y_test)).round(4))
```

```python
# cross-validation

for m in models:
    scores = cross_val_score(m, pd.concat([x_train2, x_test2]), pd.concat([y_train, y_test]), cv=5, scoring='neg_root_mean_squared_error')
    print(m)
    print(scores)
    print('Mean=', scores.mean())
    print('Std=', scores.std())
    print()
```
